```python
# Step 1: Import libraries
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
```

```python
# Step 2: Load the dataset
digits = datasets.load_digits()
```

```python
print("Dataset loaded successfully ✅")
print("Shape of data:", digits.data.shape)
print("Total images:", len(digits.images))
```

```python
# Step 3: Visualize some samples
plt.figure(figsize=(8,4))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(digits.images[i], cmap='gray')
    plt.title(f"Label: {digits.target[i]}")
    plt.axis('off')
plt.suptitle("Sample Handwritten Digits", fontsize=14)
plt.show()
```

```python
# Step 4: Prepare features and target
X = digits.data
y = digits.target
```

```python
# Step 5: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Step 6: Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Step 7: Train SVM model
svm_model = SVC(kernel='rbf', C=10, gamma=0.001)
svm_model.fit(X_train, y_train)
```

```python
# Step 8: Make predictions
y_pred = svm_model.predict(X_test)
```

```python
# Step 9: Evaluate model performance
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

```python
# Step 10: Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='d')
plt.title("Confusion Matrix — SVM Digit Classifier")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
# Step 11: Visualize Predictions
plt.figure(figsize=(8,4))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_test[i].reshape(8,8), cmap='gray')
    plt.title(f"Pred: {y_pred[i]}")
    plt.axis('off')
plt.suptitle("Sample Predictions by SVM Model", fontsize=14)
plt.show()
```
